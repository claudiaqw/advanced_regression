```python
import pandas as pd
from os import listdir
from os.path import isfile, join
from datetime import datetime
```

```python
path = "data/csse_covid_19_daily_reports_us"
l = []

files = [f for f in listdir(path) if (isfile(join(path, f)) and f.endswith('.csv'))]
for f in files:
    filename = join(path, f)
    df = pd.read_csv(filename)
    df['Date'] = (datetime.strptime(f[0:10], "%m-%d-%Y")).strftime("%Y-%m-%d")
    df = df.dropna(how='all')
    l.append(df)
```

```python
data = pd.concat(l)
data = data[data['Province_State'] != 'Diamond Princess']
data = data[data['Province_State'] != 'Grand Princess']
data = data[data['Province_State'] != 'American Samoa']
data = data[data['Province_State'] != 'Guam']
data = data[data['Province_State'] != 'Northern Mariana Islands']
data = data[data['Province_State'] != 'Puerto Rico']
data = data[data['Province_State'] != 'Virgin Islands']
data = data[data['Province_State'] != 'Recovered']
```
